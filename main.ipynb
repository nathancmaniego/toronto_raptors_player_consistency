```python
from nba_api.stats.endpoints import commonteamroster

# Toronto Raptors Team ID and the current season
team_id = 1610612761  # Toronto Raptors
season = '2024-25'    # Use the current NBA season in "YYYY-YY" format

# Fetch roster for the current season
team_roster = commonteamroster.CommonTeamRoster(team_id=team_id, season=season)
roster_df = team_roster.get_data_frames()[0]

roster_df
```

```python
roster_df = roster_df.drop(columns="HOW_ACQUIRED")
```

```python
roster_df # removed how acquired
```

Next, we will do itterows() to iterate over each row in the data frame

```python
from nba_api.stats.endpoints import playergamelog
```

Testing on Lebron James

```python
gamelog = playergamelog.PlayerGameLog(player_id=2544) #testing on lebron
data = gamelog.get_data_frames()[0]
print(len(gamelog.get_data_frames()))
data
```

```python
player_data = []
for index, row in roster_df.iterrows():
    player_id = row["PLAYER_ID"]
    player_name = row["PLAYER"]
    try:
        gamelog = playergamelog.PlayerGameLog(player_id=player_id)
        data = gamelog.get_data_frames()[0]
        if not data.empty:
            data['PLAYER'] = player_name
            player_data.append(data)
    except Exception as e:
        print(f"Error fetching data for {player_name}: {e}")
toronto_raptors_data = pd.concat(player_data, ignore_index=True)
toronto_raptors_data
```

```python
# Creating a new dataframe
consistency_data = toronto_raptors_data.groupby('PLAYER').agg(
    GAMES_PLAYED=('Game_ID', 'count'),
    PTS_MEAN=('PTS', 'mean'),
    PTS_STD=('PTS', 'std'))
consistency_data['CONSISTENCY'] = consistency_data['PTS_STD'] / consistency_data['PTS_MEAN']
consistency_data = consistency_data.sort_values(by='CONSISTENCY', ascending=True)
```
